### Loading dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/msh209/INT3625/main/data/gapminder.tsv', sep='\t')
```

```python
df.head()
```

**Q1.** Create a plot to show the relationship between life expectancy and GDP in the UK. To show a linear relationship you can use a regplot.

```python
df_uk=df[df['country']=='United Kingdom']
ax = sns.regplot(x='lifeExp', y='gdpPercap', data=df_uk)
ax.set_title('Life Expectency & GDP per capita');
ax.set_xlabel('GDP per capita');
ax.set_ylabel('Life Expectency');
```

**Q2.** Plot number of observations for each continent in 1980 and after where the population of the country was more than 10,000,000

Tip: we need to create a bar chart (countplot)

```python
ax = sns.countplot(x='continent', data=df[(df['year']>=1985) & (df['pop']>10000000)])
ax.set_title('Observations in different countries');
ax.set_xlabel('Continent');
ax.set_ylabel('Observations');
```

**Q3.**  Plot the changes of life expectancy for each contintent over different years

Try this using scatterplot and lineplot

```python
sns.scatterplot(data = df,x="year", y="lifeExp", hue='continent')
```

```python
sns.lineplot(data = df,x="year", y="lifeExp", hue='continent')
```

**Q4.** Plot the distribution of lifeExp for years 2000 onwards and different continents using facetgrid

```python
facet = sns.FacetGrid(df[df['year']>2000], col='year', hue='continent')
facet = facet.map(sns.histplot, 'lifeExp')
facet = facet.add_legend()
```

```python
facet = sns.FacetGrid(df[df['year']>2000], col='continent', hue='year')
facet = facet.map(sns.histplot, 'lifeExp')
facet = facet.add_legend()
```

**Q5.** Compare life expecty of different continents using boxplot

```python
ax = sns.boxplot(x='continent', y='lifeExp', data=df)
ax.set_title('Boxplot of life exp.')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```
