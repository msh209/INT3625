## Conditionals

**Cooking analogue - seasoning sauce**

Typically, in order to season your sauce you would proceed as follows:

* Taste sauce
* If it isn't salty enough, add salt
* If it is too salty, re evaluate your cooking skills
* If it is just fine, you don't need to do anything. Good job!

In the above, different actions need to be taken based on the outcome of the tasting process.

**Programming**

Similarly, in programming, the flow of the code might need to be altered based on the outcome of the evaluation of some condition. This can be facilitated by conditional statements, which allow to execute different sequences of commands, based on the outcome of the a evaluation of some or several conditions.

### Conditional statements in Python

**<div class="alert alert-block alert-info">
KEY POINT - The general syntax for a conditional statement in Python is:**
</div>

```Python
    if condition:
        commands
    else:
        alternativeCommands
```

**<div class="alert alert-block alert-info">
KEY POINT - Note the colons after each if/else/elif line. These are very easy to forget!**
</div>

In the above:

* `condition` is the outcome of some logical operation. It should assume a Boolean value
* `commands` is a series of commands to be executed if the condition is True 
* `alternativeCommands` is series of commands to be executed if the condition is false
* A block structure is used:
  - After the initial line all commands are indented
  - All indented commands are part of the block
  
This kind of syntax is fast to code, since it does not require any additional command to signify the end of the block, but it is also easy to mess up! For instance, if you forget to delete the spaces after you have typed all the commands in the statement, commands that you might want to add after the statement will also become part of the statement and their execution will depend on the whether or not the condition is applied!


**<div class="alert alert-block alert-success">
EXAMPLE - Here is an example of a conditional statement in action:**
</div>


````python
#Test whether a number is odd or even

a=11

if a%2==0:             # Here a%2 returns the remainder of dividing a by two. Don't forget the colon! 
    print('Even')      # What we want to happen if the above is true. Don't forget the indentation!
else:                  # Don't forget the colon! 
    print('Odd')      
````

**<div class="alert alert-block alert-danger">
EXERCISE - Don't ignore the comments - there's a task above!**
</div>

For more complicated cases, the following syntax can be used:

```Python
    if condition1:
        commands1
    elif condition2:
        commands2
    else:
        alternativeCommands
```

In this case execution would proceed as follows:

* `condition1` is tested
* if True `commands1` are executed
* if False `condition2` is tested
* if True `commands2` are executed
* if False `alternativeCommands` are executed

Also, `elif` can be repeated as many times as necessary.


**<div class="alert alert-block alert-success">
EXAMPLE - For example:**
</div>

````python
# Print at what level beef is cooked based on its temperature in Celcius degrees
# These values are not to be relied on - cook at your own risk!

beefTemperature = 47

if beefTemperature<45:
    print('Uncooked')
elif (beefTemperature>=45) and (beefTemperature<50):
    print('Rare')
elif (beefTemperature>=50) and (beefTemperature<60):
    print('Medium-rare')
elif (beefTemperature>=60) and (beefTemperature<65):
    print('Medium')
elif (beefTemperature>=65) and (beefTemperature<70):
    print('Medium-well done')
else:
    print('Well done')
    
````

```Python
    if condition1:
        if additionalCondition:
            commands1a
        else:

    elif condition2:
        commands2
    else:
        alternativeCommands
```

**Programming**

Similarly in programming there are often series of commands that need to be repeated a known or unknown number of times. This can be accomplished through the use of loops. Loops typically consist of two elements:

* Some mechanism of specifying the required number of repetitions, this could be:
  - A fixed number or set, for example 9 cuts along the onion length
  - Some condition, whose satisfaction causes the loop to exit. For example, 'Is there any onion left to chop?'
* A set of commands to be repeated, for example 'Move knife 5mm to the right'

### `break` and `continue`

In some cases, we might want to exit a loop prematurely if certain conditions apply. This can be accomplished using the `break` command and the following syntax:

```Python
    for item in sequence:
        commands
        
        if condition:
            break
```

This will execute the loop normally but will exit if at some point the `condition` in the conditional statement is met.

**<div class="alert alert-block alert-success">
EXAMPLE - For example:**
</div>

````python
a=10           #assign a number to a, try for example the values 100 and 10000

#Half the value of the variable 'a' ten times but stop if its value is less than one

for i in range(0,10):  #start a loop with 10 iterations
    print('i=',i)      #print i to keep track of the number of iterations
    a /= 2             #divide a by 2
    
    if a<1:            #add conditional statement to check whether a is less than 1
        #Print a statement if the condition is met
        print('Condition met, exiting loop prematurely')
        break          #use break to exit the loop
        
print(a)
````

In other cases, we might want some of the commands in a loop to be executed only if certain conditions apply. This can be accomplished using the `continue` command and the following syntax:

```Python
    for item in sequence:
        commands
        
        if condition:
            continue
        
        moreCommands
```

This will skip the remaining commands in the loop (`moreCommands`) if the `condition` is met.

**<div class="alert alert-block alert-success">
    EXAMPLE - For example:** 
</div>

````python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'] #list with all days of the week

#loop days of the week
for d in days:
    #print all days
    print(d, 'is a day of the week')
    
    #Check if the day is a holiday
    if d in ['Sunday','Saturday']:
        continue #if the day is a holiday continue (i.e. go back to the start of the loop)
    
    #print working days
    print(d, 'is a working day :(')
````

The factorial of an integer number $n$ is denoted as $n!$ and can be computed as:

$n! = 1 \cdot 2 \cdot 3 \cdot \dots \left( n-2 \right) \cdot \left( n-1 \right) \cdot n$

**Problem Description**

The Collatz Conjecture is an open research problem in mathematics. It states that the following iterated method will always reduce a number to one:

* If $n$ is odd, multiply by three and add one.
* If $n$ is even, divide by two.

For instance if we start with $n = 13$, we multiply by three and add one to get $n = 40$, repeating this:

* 40/2 = 20
* 10/2 = 5
* 5 * 3 + 1 = 16
* 16/2 = 8
* 8/2 = 4
* 4/2 = 2
* 2/2 = 1

and we have reduced to one. More information can be found in [Wikipedia](https://en.wikipedia.org/wiki/Collatz_conjecture), as well as this [video](https://www.youtube.com/watch?v=094y1Z2wpJg).
